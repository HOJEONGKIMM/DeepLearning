```python
# touch, donut, sound, frame, glass, elbow, hello, smile, night, imply, zebra -> 11개
import numpy as np
data = np.array(
    [
        [t,o,u,c,h],
        [d,o,n,u,t],
        [s,o,u,n,d],
        [f,r,a,m,e],
        [g,l,a,s,s],
        [e,l,b,o,w],
        [h,e,l,l,o],
        [s,m,i,l,e],
        [n,i,g,h,t],
        [i,m,p,l,y],
        [z,e,b,r,a]
    ]
)
data.shape
```

```python
# 시계열 데이터 : 데이터가 순서를 가지고 있다. (보통 3차원의 데이터)
# sample, timestep, feature
# sample : 가지고 있는 데이터의 개수 (11)
# timestep : 하나의 데이터가 가지고 있는 순서 (5)
# feature : 하나의 timestep이 표현할 수 있는 개수 (26)
data.shape
```

```python
# 앞의 4글자는 문제 데이터
# 뒤의 1글자는 정답 데이터
X = data[:, :-1]
y = data[:, -1]
X.shape, y.shape
```

```python
# 시계열 딥러닝 모델
# MLP, CNN, YOLO -> 여러 데이터를 학습할 수 있음
#                -> 순서에 따라 의미가 달라지는 데이터는 학습은 되지만, 결과가 나오지 않음
#                -> 순서를 상관하지 않고 학습
# 시계열 딥러닝 모델
# RNN, LSTM -> 여러 데이터를 학습할 수 있음
#           -> 순서에 따라 의미가 달라지는 데이터 학습도 되고, 결과도 잘 나옴
#           -> 순서를 기억한 채로 학습
```

```python
from os import access
# 모델링
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

model = Sequential()
# input_shape : 입력 데이터(문제, X)의 크기(개수는 상관하지 않는다, 데이터 중 1개를 의미)
# X 데이터는 앞의 4개 글자, y 데이터는 뒤의 1개 글자
# activation = 'tanh' : 임팩트 있는 값들은 기억하지만 아닌 것들은 잊는다
model.add(SimpleRNN(units=30, input_shape = (4,26), activation= 'tanh'))

# units = 26 : 알파벳의 개수 -> 다중분류
model.add(Dense(units = 26, activation = 'softmax'))

# 학습 방법 설정
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

# 학습
model.fit(X,y,epochs=300)
```

```python
# touch, donut, sound, frame, glass, elbow, hello, smile, night, imply, zebra
model.predict([[s,o,u,n]])
```
